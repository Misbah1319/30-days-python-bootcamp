```python
# Project 1 - Day 21
# Goal: Find the most repetitive element in a list

# Step 1: Create the list
L1 = [1, 2, 2, 3, 2, 3, 4, 5]  # The list given in the project

# Step 2: Use a dictionary to count the frequency of each element
frequency = {}  # Empty dictionary to store counts

for item in L1:  # Loop through each element in the list
    if item in frequency:  # If element is already in dictionary
        frequency[item] += 1  # Increment its count
    else:
        frequency[item] = 1  # If not, initialize count to 1

# Step 3: Find the element with the highest count
most_repetitive = max(frequency, key=frequency.get)
# max() finds the key with maximum value in dictionary

# Step 4: Print the result
print("The most repetitive element is:", most_repetitive)
```
